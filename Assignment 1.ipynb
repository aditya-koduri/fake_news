```python
df_fake = df[df['fake']==1]

df2 = df_fake['text'].apply(lambda x : x.split())

list_all_words = df2.tolist()

full_list = []

for i in range(0,len(list_all_words)) :
    for j in range(0,len(list_all_words[i])) :
        full_list.append(list_all_words[i][j])


counts = Counter(full_list)

counts_df = pd.DataFrame.from_dict(counts, orient='index').reset_index()
counts_df.columns = ['word','count']

copy_data = counts_df.copy()
with pd.ExcelWriter(r"C:\Users\adity\Desktop\Columbia\Spring 2021\DS_PP\Assignment\words_hist_f.xlsx") as writer:  
    copy_data.to_excel(writer, sheet_name='fake_words')
```
